```python
import sqlite3 as lite
import sys
import pandas as pd
import numpy as np
import os
con = lite.connect('Tweets.db')

with con:
    
    cur = con.cursor()    
    cur.execute('SELECT SQLITE_VERSION()')
    
    data = cur.fetchone()
    
    print ("SQLite version: %s" % data)
```

```python
with con:
    
    cur = con.cursor()    
    
    cur.execute("DROP TABLE IF EXISTS Tweetss")
    cur.execute("CREATE TABLE Tweetss(ID INT, tweet TEXT, retweet INT)")
    cur.executemany("INSERT INTO Tweetss VALUES(?,?,?)", list_tweets)
```

```python
with con:
    
    con.row_factory = lite.Row
    
    cur = con.cursor() 
    cur.execute("SELECT * FROM Tweetss")

    rows = cur.fetchall()

    for row in rows:
        print ("%s   %s   %s" % (row["ID"],row["tweet"], row["retweet"]))
```
